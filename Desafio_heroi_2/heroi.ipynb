```python
import random
import sys

class Jogo():
    def __init__(self):
        self.cont_JOGADOR = 0
        self.cont_CPU = 0
        self.nome = input('Digite o nome do seu heroi: ')
        print(f'O herói {self.nome} foi criado. Que comecem os jogos.')

    def dados(self):
        print("Vamos rodar os dados")
        self.CPU = random.randint(1, 6)
        print(f'O computador tirou {self.CPU}')
        self.JOGADOR = random.randint(1, 6)
        print(f'{self.nome} tirou {self.JOGADOR}')

    def regra(self):
        if self.CPU > self.JOGADOR:
            print("O computador Ganhou")
            self.cont_CPU += 1
        else:
            print(f"O {self.nome} ganhou essa")
            self.cont_JOGADOR += 1

    def resultado(self):
        if self.cont_JOGADOR <=10:
            xp_log = "Ferro"
        elif self.cont_JOGADOR >= 11 and self.cont_JOGADOR <= 20:
            xp_log = "Bronze"
        elif self.cont_JOGADOR >= 21 and self.cont_JOGADOR <= 50:
            xp_log = "Prata"
        elif self.cont_JOGADOR >= 51 and self.cont_JOGADOR <= 80:
            xp_log = "Ouro"
        elif self.cont_JOGADOR >= 81 and self.cont_JOGADOR <= 90:
            xp_log = "Diamante"
        elif self.cont_JOGADOR >= 91 and self.cont_JOGADOR <= 100:
            xp_log = "Lendário"
        elif self.cont_JOGADOR >= 101:
            xp_log = "Imortal"
        print(f'O herói {self.nome} tem o saldo de {self.cont_JOGADOR} e está no nível {xp_log}')

    def final(self):
        final = input('Deseja jogar mais uma partida? S/N ')
        if final.upper() == "N":
            print(f'Fim do jogo. Até a próxima {self.nome}')
            return False  # Retorna False para indicar que o jogo deve parar
        else:
            return True  # Retorna True para indicar que o jogo deve continuar

jogo = Jogo()
while True:
    jogo.dados()
    jogo.regra()
    jogo.resultado()
    if not jogo.final():
        break
```
